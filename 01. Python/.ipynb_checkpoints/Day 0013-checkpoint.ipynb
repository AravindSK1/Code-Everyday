For n between 1 to 1000, list all values where a^3 + b^3 = c^3 + d^3

```python
import numpy as np
import pandas as pd
```

```python
def matching_cubes_sum(m):
    n = m + 1
    c_d = {}
    c_cube_d_cube = {}

    ## Key = c**3 + d**3
    ## value = [c,d]
    for c in range(1,n):
        for d in range(1,n):
            arithematic_expression = c**3+d**3
            # if the key value is already present return the value assigned, else 0
            c_cube_d_cube[arithematic_expression] = c_cube_d_cube.get(arithematic_expression,0) + 1

            # Maintain a value dictionary to capture c and d
            if arithematic_expression in c_d:
                c_d[arithematic_expression].append(c)
                c_d[arithematic_expression].append(d)
            else:
                c_d[arithematic_expression] = [c,d]

    matching_cd_combination = [c_d[cube_sum_value] for cube_sum_value, count in c_cube_d_cube.items() if count != 1]

    return matching_cd_combination
```

```python
matching_cubes_sum(100)
```
