```python
class linearRegression:
    # 1. Constructor for epochs and learning rate
    def __init__(self):
        self.epochs = 10001
        self.learning_rate = 0.005
    
    # calculate the predicted value
    def prediction(self, w, X):
        y_hat = np.dot(w.T, X)
        return y_hat
    
    # calculate cost
    def cost(self, y, y_hat):
        MSE = 1/self.n * np.sum(np.power((y-y_hat),2))
        return MSE
    
    # calculate gradients
    def gradient_descent(self, w, X, y, y_hat):
        dw0_dw = -2/self.n * (np.dot(X, (y-y_hat).T))
        w = w - self.learning_rate * dw0_dw
        return w
    
    def shape(self, X):
        self.k, self.n = X.shape
        
    # fit the model
    def fit(self, X, y):
        # get the shape of x
        self.shape(X)
        
        # add 1 for bias
        ones = np.ones((1,self.n))
        X    = np.append(ones, X, axis = 0) 
       
        # update shape
        self.shape(X)
        
        # initialize weights
        w    = np.zeros((self.k),)
        
        # iterations
        for epoch in range(1, self.epochs):
            y_pred = self.prediction(w, X)
            C      = self.cost(y, y_pred)
            w      = self.gradient_descent(w, X, y, y_pred)
            
            if epoch % 100 == 0:
                print(f"Cost at iteration {epoch} is {C}")
                
        return y_pred, w
    
if __name__ == "__main__":
    np.random.seed(123)
    X = np.random.rand(1,500)
    y = np.random.randint(1,7) * X + np.random.randint(1,10) + np.random.randn(1,500) * 0.1
    
    lr = linearRegression()
    y_pred, w = lr.fit(X,y)
```
